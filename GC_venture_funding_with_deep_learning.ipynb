## Prepare the data to be used on a neural network model

### Step 1: Read the `applicants_data.csv` file into a Pandas DataFrame. Review the DataFrame, looking for categorical variables that will need to be encoded, as well as columns that could eventually define your features and target variables.  


```python
# Upload credit_card_transactions.csv to Colab
# Here is some slick code to upload csv files into Google Colab.
from google.colab import files

csv_file = files.upload()
```

```python
# Read the applicants_data.csv file from the Resources folder into a Pandas DataFrame
# Here we are having the pandas read_csv() function read in the csv file and create dataframe with it. We pass the name of the csv file as an argument.
applicant_data_df = pd.read_csv(Path('applicants_data.csv'))

# Review the DataFrame
# Here we call the name of the dataframe we just created to review and see what the data is in the dataframe.
applicant_data_df
```

```python
# Review the data types associated with the columns
# Here we can use the dtypes function to see what kind of data we have in our dataframe. This is very useful to see because we are building a Neural Network later
# and its going to benefit us and the accuracy of the NN to change the 'objects' into integers. Thats becasue the algorithims cannot make sense of different categorical
# labels. We will assign the categories an integer.
applicant_data_df.dtypes
```

### Step 2: Drop the “EIN” (Employer Identification Number) and “NAME” columns from the DataFrame, because they are not relevant to the binary classification model.

```python
# Drop the 'EIN' and 'NAME' columns from the DataFrame
# Here we drop the 'EIN' and 'NAME' columns becasue they really serve us no purpose in the neural network model. In fact the 'EIN' could break the model, and the 
# 'Name' is useless as well as being categorical. So we just drop them useing the .drop() function.
applicant_data_df = applicant_data_df.drop(columns=['EIN', 'NAME'])

# Review the DataFrame
# Here we can review the dataframe to see if indeed we dropped the two columns.
applicant_data_df
```

### Step 3: Encode the dataset’s categorical variables using `OneHotEncoder`, and then place the encoded variables into a new DataFrame.

```python
# Create a list of categorical variables 
# Here we are making a list of all the categorical columns. These categories can break the model unless we modify them into integers.
categorical_variables = list(applicant_data_df.dtypes[applicant_data_df.dtypes == "object"].index)

# Display the categorical variables list
# A simple display() command to view our list.
display(categorical_variables)
```

```python
# Create a OneHotEncoder instance
# Here we are going to use OneHotEncoder to cfonvert all the categories into a number value. Basically what happens is the sklearn.preprocessing module
# is able to give these categorical labels a number value either 1 or 0, and depending on how many cateories there is, relates to how many new columns 
# are created. So for example if there was previously 4 'Organization' categories, OneHotEncoder would create 4 new columns in the dataframe and give each row
# one value depending on their category.. Here we create the OneHotEncoder instance and give it a new name = enc.
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse=False)
```

```python
# Encode the categorcal variables using OneHotEncoder
# Here we use the .fit_transform() function and pass it the categorical_variables columns from the dataframe.
encoded_data = enc.fit_transform(applicant_data_df[categorical_variables])
```

```python
# Create a DataFrame with the encoded variables
# Here we are creating a new dataframe that will include the new columns the OneHotEncoder created earlier, so were previously we had 10 columns, now 
# we have 114 columns. To do this we are using pandas DataFrame() function as well as sklearn.preprocessing .get_feature_names_out() function.
encoded_df = pd.DataFrame(encoded_data, columns = enc.get_feature_names_out(categorical_variables))


# Review the DataFrame
# And here we call on the dataframe to review it.
encoded_df
```

### Step 4: Add the original DataFrame’s numerical variables to the DataFrame containing the encoded variables.

> **Note** To complete this step, you will employ the Pandas `concat()` function that was introduced earlier in this course. 

```python
# Add the numerical variables from the original DataFrame to the one-hot encoding DataFrame
# Here we use the pd.concat() function to join the two dataframes, encoded_df and the applicant_data_df dataframes. I only wanted to include the 
# the numerical variables from the original dataframe called applicants_data_df, because the categorical columns have already been included into a new
# dataframe called encoded_df. So when I concat, I call the applicants_data_df dataframe but only include the columns I want to include so we don't
# double up on the data, or use the same data twice.

encoded_df = pd.concat([encoded_df, applicant_data_df[['STATUS', 'ASK_AMT', 'IS_SUCCESSFUL']]], axis=1)

# Review the Dataframe
encoded_df
```

### Step 5: Using the preprocessed data, create the features (`X`) and target (`y`) datasets. The target dataset should be defined by the preprocessed DataFrame column “IS_SUCCESSFUL”. The remaining columns should define the features dataset. 



```python
# Define the target set y using the IS_SUCCESSFUL column
# Here we are taking the colume 'IS_SUCCESSFUL' from the dataframe and setting it to y, for the NN model we are getting ready to create. This 
# column will be aour target column, or the column that we are trying to predict with the model we will build.
y = encoded_df['IS_SUCCESSFUL']

# Display a sample of y
# And here is a quick display to see what y looks like. We just call the first 5 rows.
y
```

```python
# Define features set X by selecting all columns but IS_SUCCESSFUL
# Here we want all the columns except 'IS_SUCCESSFUL', so we use the drop() and pass it the column we want to drop. These columns will make up the features
# in our test and train dataset for the prediction model that we will build. We assign this dataset to X. Basically this is the dataset that the NN
# model will use to make predictions.
X = encoded_df.drop(columns=['IS_SUCCESSFUL'])

# Review the features DataFrame
# Here we review what X will look like. Note, it does not include the 'IS_SUCCESSFUL' column.
X[:5]
```

### Step 6: Split the features and target sets into training and testing datasets.


```python
# Split the preprocessed data into a training and testing dataset
# Assign the function a random_state equal to 1
# Here we use the train_test_split() function from sklearn.model_selection. This function assigns some data to the training component of the 
# model and reserves some for the test component of the model. Random_state=1 has to do with how the function selects randomly from the data to assign it
# to either test or train.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

### Step 7: Use scikit-learn's `StandardScaler` to scale the features data.

```python
# Create a StandardScaler instance
# Here we are using the StandardScaler() function to scale all the numerical values in the dataset. The purpose is so that all the valuse are scaled, 
# specifically the 'ASK_AMOUNT' column will now be scaled to the rest of the data and that column wont screw up the model. This is very important for
# a robust accurate model. So we create a instance for the StandardScaler and assign it to the variable scaler.
scaler = StandardScaler()

# Fit the scaler to the features training dataset
# And here we pass the scaler instance to the fit() function with X_train dataset as the argument.
X_scaler = scaler.fit(X_train)

# Fit the scaler to the features training dataset
# Here we use the transform() function and pass it to X_scaler with the relevant datasets. Now we have scaled datasets to use in our model.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile and Evaluate a Binary Classification Model Using a Neural Network

### Step 1: Create a deep neural network by assigning the number of input features, the number of layers, and the number of neurons on each layer using Tensorflow’s Keras.

> **Hint** You can start with a two-layer deep neural network model that uses the `relu` activation function for both layers.


```python
# Define the the number of inputs (features) to the model
# Here we are using the len() function and passing the X_train dataset , row 0 as the argument, to see how many columns we can assign to number_input_features.
number_input_features = len(X_train.iloc[0])

# Review the number of features
# Here we review the number_input_features and see it has 116 rows, wich adds up becasue after the OneHotEncoding we had 117 less 1 column for the target.
number_input_features
```

```python
# Define the number of neurons in the output layer
# Here we are defining the number of neurons in the output layer of the model. We use 1, becasue there is always just 1 output layer in a NN. It is
# responsible for producing the final output, and takes in  the inputs which are passed in from the layers before it, performs the calculations via
# its neurons and then the output is computed.
number_output_neurons = 1
```

```python
# Define the number of hidden nodes for the first hidden layer
# Here we are defining the number of hidden nodes in the first hidden layer, its common practice to basically use half the number of input_features 
# for the first hidden layer. So we add the input_features and ouput and divide by 2 to get the number of nodes in the first hidden layer of our model.
hidden_nodes_layer1 =  (number_input_features + 1) // 2

# Review the number hidden nodes in the first layer
# Here we review how many nodes in the first hidden layer
hidden_nodes_layer1
```

```python
# Define the number of hidden nodes for the second hidden layer
# The nodes for the second hidden layer should be around half the nodes for the first hidden layer, so we do a similiar equation to find that number.
hidden_nodes_layer2 =  (hidden_nodes_layer1 + 1) // 2

# Review the number hidden nodes in the second layer
# Here we review how many nodes in the second hidden layer
hidden_nodes_layer2
```

```python
# Create the Sequential model instance
# Here we create an instance for the Sequential model and assign it to the variable nn.
nn = Sequential()
```

```python
# Add the first hidden layer
# Here we use the add() function to add the first hidden layer to nn, and define some of the parameters, specifically, how many units, input_dimension
# and the activation function. We have chosen to use the rectified linear unit (ReLU) function (Links to an external site.). It is currently the world's 
# most-used activation function when it comes to training deep neural networks.
nn.add(Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation='relu'))
```

```python
# Add the second hidden layer
# Here we use the add() function to add the second hidden layer to nn, and define some of the parameters, specifically, how many units
# and the activation function. We have chosen to use the rectified linear unit (ReLU) function (Links to an external site.). It is currently the world's 
# most-used activation function when it comes to training deep neural networks. 
nn.add(Dense(units=hidden_nodes_layer2, activation='relu'))
```

```python
# Add the output layer to the model specifying the number of output neurons and activation function
# Here we use the .add() function to add the parameters of the output layer to the nn model. Now, for
# our classification model’s output layer, we use the sigmoid activation function. The sigmoid function
# will transform the output to a range between 0 and 1. 
nn.add(Dense(units=number_output_neurons, activation='sigmoid'))
```

```python
# Display the Sequential model summary
# Here we use the summary() function to display the structure of the model, the output shape and number of
# weights—of each layer in our neuron model.
nn.summary()
```

### Step 2: Compile and fit the model using the `binary_crossentropy` loss function, the `adam` optimizer, and the `accuracy` evaluation metric.


```python
# Compile the Sequential model
# Here we compile the model by using the compile() function provided by keras, and pass several parameters. Where previously we designed the model
# and defined the parameters, here with the compile() function we are actually going to construct the model. We include loss, optimizer and metrics 
# as parameters. The loss function uses machine learning algorithms to score the performance of the model after each of these iterations. This allows
# us to see how the model’s performance changes over each iteration. We also use the optimizer function, this function shapes and molds a neural
# network model while the model is trained on the data, to ensure that the model performs to the best of its ability. At a glance, an optimization 
# function reduces the model’s losses and provides the most accurate output possible. The metrics parameter specifies additional metrics that assess 
# the quality of a neural network model.
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model using 50 epochs and the training data
# Here we use the .fit() function to train the nn model to X_train_scaled dataset and y_train dataset, also we set the amount of epochs to 50.
# An epoch is a single pass of the entire training dataset through the model.
fit_model = nn.fit(X_train_scaled, y_train, epochs=50)
```

### Step 3: Evaluate the model using the test data to determine the model’s loss and accuracy.


```python
# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
# Now we can evaluate the model by using the .evaluate() function. This function shows the performance of the model.
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)
# Display the model loss and accuracy results
# Here we use a print command to print a f' string 
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

### Step 4: Save and export your model to an HDF5 file, and name the file `AlphabetSoup.h5`. 


```python
# Set the model's file path
# Here we name the model AlphabetSoup.h5 and give it the variable name file_path.
file_path = 'AlphabetSoup.h5'

# Export your model to a HDF5 file
# Here we use the save() function to save the model to the file path name we preciously determined. This way we can pass the model to a freind or
# colleague.
nn.save(file_path)
```
